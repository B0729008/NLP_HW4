```python
#簡轉繁
from opencc import OpenCC
cc = OpenCC('s2twp')
```

```python
# -*- coding:utf8 -*-
import os
import jieba
import json
import simplejson

def splitSentence(inputFile):
    fin = open(inputFile, 'r',encoding="utf-8")
    words=[]
    for eachLine in fin:
        eachLine_dict = simplejson.loads(eachLine)
        #print(eachLine_dict['text'])
        line = eachLine_dict['text'].strip()
        line=line.strip('\n')                     
        wordList = list(jieba.cut(line))  
        outStr = []
        for word in wordList:
            if len(word)>1:
                temp=cc.convert(word)
                outStr.append(temp)
        words.append(outStr)
    fin.close()
    return words
```

```python
import os
fout = open('mydict.txt', 'w',encoding='utf-8')
path = "wiki_zh" #資料夾目錄
folder1= os.listdir(path) #得到資料夾下的所有檔名稱
document=[]

for folder in folder1:
    files= os.listdir(path+"/"+folder)
    for file in files:
        fin = open(path+"/"+folder+"/"+file, 'r',encoding="utf-8")
        for eachLine in fin:
            eachLine_dict = simplejson.loads(eachLine)
            #print(eachLine_dict['text'])
            a=cc.convert(eachLine_dict['title'])
            fout.write(a+'\n')
        fin.close()
fout.close()
```

```python
jieba.load_userdict('mydict.txt')
jieba.add_word('李知恩')
for folder in folder1:
    files= os.listdir(path+"/"+folder)
    for file in files:
        print("Now input file:"+folder+"/"+file)
        document+=splitSentence(path+"/"+folder+"/"+file)
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s',level=logging.INFO)
                     
#min_word_count = 40   # Minimum word count                        
num_workers = 4       # Number of threads to run in parallel 

from gensim.models import word2vec
print("Training model...")
model = word2vec.Word2Vec(document, vector_size=300, workers=num_workers , sg=1, epochs=5)
model.init_sims(replace=True)
model_name = "40minwords_10epochs"
model.save(model_name)
```

```python
result=model.wv.most_similar('李知恩', topn=20)
f=open('HW4_Result','w',encoding='utf-8')
print(result, file=f)
result
```
